```python
import pandas as pd
import os
import numpy as np
import seaborn as sns
import sys
sys.path.append('./..')
import glob
import matplotlib.pyplot as plt
```

```python
DIR = 'us_import1'
```

```python
def _plot_(DIR):
    df1 = pd.read_csv('./classifier_op/KNN/{}/results.csv'.format(DIR), index_col=None)
    prec1 = df1.groupby('idx').mean().reset_index()['precision'].values

    df2 = pd.read_csv('./classifier_op/DT/{}/results.csv'.format(DIR), index_col=None)
    prec2 = df2.groupby('idx').mean().reset_index()['precision'].values

    file = list(sorted(glob.glob(os.path.join('./../hitl_model_2/Output', DIR, 'results_withInput**.csv'))))[-1]
    df3 = pd.read_csv(file , index_col=None)
    prec3 = df3.groupby('idx').mean().reset_index()['Prec@next_20'].values
    sns.set_context("paper", rc={"font.size":15,"axes.titlesize":14,"axes.labelsize":14})  
    
    prec1 = np.insert(prec1, 0, [0.5])
    prec2 = np.insert(prec2, 0, [0.5])
    prec3 = np.insert(prec3, 0, [0.5])

    fig, axs = plt.subplots(nrows=1, ncols=1, figsize=[6.5,4.5])
    sns.set_style("whitegrid")
    plt.ylim(0, 1.01)
    x = np.arange(1, prec1.shape[0]+1)
    g = sns.lineplot( x=x, y=prec1, markers=True,color='red', label='Baseline 1')
    x = np.arange(1,prec2.shape[0]+1)
    g = sns.lineplot( x=x, y=prec2, markers=True,color='blue', label='Baseline 2')
    x = np.arange(1,prec3.shape[0]+1)
    g = sns.lineplot( x=x, y=prec3, markers=True, color='green', linewidth= 3, label='Linear Scoring Model')
    g.axhline(0.5, linestyle ='-.',color='orange',label='E(precision) , no input')
    axs.set(xlabel='Batch')
    plt.legend(fontsize=14)
    plt.xlabel('Batch',fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.tight_layout()

    plt.savefig('result_prec_20_{}.pdf'.format(DIR))
    plt.show()
    plt.close()
    return


# g = sns.lineplot(data=df, x="idx", y=y_col, markers=True,  ax=axs[i][j],color='red')
#         g.axhline(0.5,color='r',label='E(precision) , no input')
#         axs[i][j].set(xlabel='Batch')
```

```python
_plot_('us_import1')
_plot_('us_import2')
_plot_('us_import4')
```
